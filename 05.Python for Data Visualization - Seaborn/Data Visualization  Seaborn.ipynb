## Distribution Plots

```python
import seaborn as sns
```

```python
%matplotlib inline
```

```python
tips = sns.load_dataset('tips') # build in dataset in seaborn library
```

```python
tips.head()
```

```python
# plot a uni-variable(one varibale of a dataset)
sns.distplot(tips['total_bill'])
```

```python
sns.distplot(tips['total_bill'], kde=False)
```

```python
# change number of bins(bar)
sns.distplot(tips['total_bill'], kde=False, bins=30)
```

```python
# combine to distplot
sns.jointplot(x='total_bill',y='tip',data=tips)# compaire the total bill to the tip size
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips, kind='hex')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips, kind='reg')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips, kind='kde')
```

```python
# pairplot to plot relationship across an entire data frame of numerical column
sns.pairplot(tips)
```

```python
sns.pairplot(tips, hue='sex') # in hue we pass a categorical column of the dataset
```

```python
sns.pairplot(tips, hue='sex', palette='coolwarm')
```

```python
sns.rugplot(tips['total_bill'])
```

```python
sns.kdeplot(tips['total_bill'])
```

# Categorical Plots
Plot categorical data

```python
# display the average or mean of tatal bill per categorical column which is sex
sns.barplot(x='sex',y='total_bill', data=tips) # x column we pass categorical data and for y column we pass numerical data
```

```python
import numpy as np
# get the standard deviation of total bill per sex as category
sns.barplot(x='sex',y='total_bill', data=tips, estimator=np.std) # estimator take a function
```

```python
# count plot: is like barplot but the estimator is counting the number of occurence
sns.countplot(x='sex', data=tips) # we can see that there is more male in the dataset than female
```

```python
# box plot and violin plot: this type of plot are use to show distribution of categorical data
sns.boxplot(x= 'day',y='total_bill', data=tips) # see the distribution of total bill per day
```

```python
sns.boxplot(x= 'day',y='total_bill', data=tips, hue='smoker')
```

```python
 sns.violinplot(x='day', y='total_bill', data=tips)
```

```python
 sns.violinplot(x='day', y='total_bill', data=tips, hue='sex')
```

```python
 sns.violinplot(x='day', y='total_bill', data=tips, hue='sex', split=True)
```

```python
 sns.stripplot(x='day', y='total_bill', data=tips)
```

```python
 sns.stripplot(x='day', y='total_bill', data=tips,jitter=True)
```

```python
 sns.stripplot(x='day', y='total_bill', data=tips, hue='sex', jitter=True)
```

```python
sns.swarmplot(x='day', y='total_bill', data=tips)
```

```python
sns.violinplot(x='day', y='total_bill', data=tips)
sns.swarmplot(x='day', y='total_bill', data=tips, color='black')
```

```python
# sns.factorplot(x='day', y='total_bill', data=tips, kind='bar') # have been rename as catplot
sns.catplot(x='day', y='total_bill', data=tips, kind='bar')
```

# Grids

```python
iris = sns.load_dataset('iris')
```

```python
iris.head()
```

```python
iris['species'].unique()
```

```python
g = sns.PairGrid(iris)
```

```python
import matplotlib.pyplot as plt
g = sns.PairGrid(iris)
g = g.map(plt.scatter)
```

```python
g = sns.PairGrid(iris)
g.map_diag(sns.distplot)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```
