### Plot Type

```python
# area plot
df2.plot.area(alpha=0.4)
```

```python
# bar plot
df2.plot.bar() # work on categorical index
df2.plot.bar(stacked=True)
```

```python
# histogram plot
df1['A'].plot.hist()
```

```python
# line plot
df1.plot.line(x= 'A', y='B', figsize=(12,3), lw=1)
```

```python
# scatter
df1.plot.scatter(x='A', y='B', c='C', cmap='coolwarm', s=df1['A']*30)
```

```python
# box plot
df1.plot.box()
```
