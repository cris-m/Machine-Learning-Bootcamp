```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0, 5, 11)
y = x**2
```

# Functional method

```python
plt.plot(x, y)
# pt.show()
```

```python
# Add title to the plot
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
plt.plot(x, y)
```

```python
# multiplot on the same canvas
plt.subplot(1, 2, 1)
plt.plot(x, y, 'r')
plt.subplot(1, 2, 2)
plt.plot(y, x, 'b')
```

# Object Oriented Method

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8]) # 0.1(top),0.1(bottom), 0.8(width), 0.8(height)
axes.plot(x, y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Title')
```

```python
# to set of plot on one canvas
fig = plt.figure()
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])
axes1.plot(x, y)
axes1.set_title('LARGER PLOT')

axes2.plot(y, x)
axes2.set_title('SMALLER PLOT')
```

```python
# sub plot
fig, axes = plt.subplots(nrows=1, ncols=2)
#axes.plot(x,y)
# deal with overlaping
#plt.tight_layout()
```

```python
axes # it is an array
```

```python
# iterates through axes
fig, axes = plt.subplots(nrows=1, ncols=2)
for current_ax in axes:
    current_ax.plot(x,y)
```

```python
fig, axs = plt.subplots(nrows=2, ncols=1, figsize=(6, 4))
axs[0].plot(x, y)
axs[1].plot(y, x)
plt.tight_layout()
```

# Special Plot Type

```python
### Scatter Plot
sct = plt.figure()
ax = sct.add_axes([0, 0, 1, 1])
ax.scatter(x, y)
```

```python
## Hystogram
from random import sample
data = sample(range(1,1000), 100)
hst = plt.figure()
ax = hst.add_axes([0, 0, 1, 1])
ax.hist(data)
```

```python
### Box plot
import numpy as np
data = [np.random.normal(0, std, 100) for std in range(1, 4)]
bx = plt.figure()
ax = bx.add_axes([0, 0, 1, 1])
ax.boxplot(data, vert=True, patch_artist=True)
```

# Advanced Matplotlib Concepts

### Logarithmic scale

```python
fig, axes = plt.subplots(1, 2, figsize=(10, 4))
axes[0].plot(x, x**2, x, np.exp(x)) # exp numpy function calculate exponential in the input array
axes[0].set_title('Normal scale')

axes[1].plot(x, x**2, x, np.exp(x))
axes[1].set_yscale('log')
axes[1].set_title('"Logarithmic scale (y)"')
```

### Placement of ticks and custom tick labels

```python
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, x**2, x, np.exp(x))
ax.set_xticks([1, 2, 3, 4, 5])
```

```python
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, x**2, x, np.exp(x))
ax.set_xticks([1, 2, 3, 4, 5])
ax.set_xticklabels([r'$\alpha$', r'$\beta$', r'$\gamma$', r'$\delta$', r'$\epsilon$'], fontsize=20)
ax.set_yticks([0, 50, 100, 150])
```

### Axis grid

```python
fig, axes = plt.subplots(1, 2, figsize=(10, 4))
axes[0].plot(x, x**2)
axes[0].set_title('X Square')
axes[0].grid(True)# default grid appearance

axes[1].plot(x, x**3)
axes[1].set_title('X Cube')
axes[1].grid(color='red', alpha=0.5, linestyle='dashed', linewidth=0.5)# customer grid appearance
```

### Twin axes

```python
fig, ax1 = plt.subplots()
ax1.plot(x, x**2, linewidth=2, color='red') 
ax1.set_ylabel(r'are $(m^2)$', fontsize=18, color='red')
for label in ax1.get_yticklabels():
    label.set_color('red')

ax2 = ax1.twinx()
    
ax2.plot(x, x**3, linewidth=2, color='blue')
ax2.set_ylabel(r'are $(m^3)$', fontsize=18, color='blue')
for label in ax2.get_yticklabels():
    label.set_color('blue')
```
